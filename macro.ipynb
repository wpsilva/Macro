Correlações

Indicadores Macro

```python
selic2 = pd.DataFrame()
confianca_consumidor2 = pd.DataFrame()
pib2 = pd.DataFrame()
incc2 = pd.DataFrame()
ipca2 = pd.DataFrame()
dolar2 = pd.DataFrame()
```

```python
selic2["SELIC"] = nasdaqdatalink.get("BCB/432", start_date="2015-01-01", collapse='monthly')
ipca2["IPCA"] = nasdaqdatalink.get("BCB/13522", start_date='2015-01-01', collapse='monthly')
incc2["INCC"] = nasdaqdatalink.get("BCB/192", start_date='2015-01-01', collapse='monthly')
pib2["PIB"] = nasdaqdatalink.get("BCB/4380", start_date='2015-01-01', collapse='monthly')
confianca_consumidor2["CONFIANCA"] = nasdaqdatalink.get("BCB/4393", start_date='2015-01-01', collapse='monthly')
dolar2["DOLAR"] = nasdaqdatalink.get("BCB/10813", start_date='2015-01-01', collapse='monthly')
```

```python
macros = pd.concat([selic2, confianca_consumidor2, pib2, ipca2, incc2, dolar2], axis=1)
macros
```

```python
matrix_correlacao = macros.corr()
```

```python
matrix_correlacao
```

```python
lista_macro = macros.columns
sns.heatmap(matrix_correlacao, cmap="coolwarm", center=0, annot=True, fmt='.1g', xticklabels=lista_macro, yticklabels=lista_macro)
```

```python
fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])
fig.add_trace(go.Scatter(name="SELIC", x=macros.index, y=macros["SELIC"]), row=1, col=1)
fig.add_trace(go.Scatter(name="IPCA", x=macros.index, y=macros["IPCA"]), row=1, col=1, secondary_y=False)
fig.show()
```

```python
fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])
fig.add_trace(go.Scatter(name="PIB", x=macros.index, y=macros["PIB"]), row=1, col=1)
fig.add_trace(go.Scatter(name="DOLAR", x=macros.index, y=macros["DOLAR"]), row=1, col=1, secondary_y=True)
fig.show()
```

```python
selic3 = pd.DataFrame()
confianca_consumidor3 = pd.DataFrame()
pib3 = pd.DataFrame()
incc3 = pd.DataFrame()
ipca3 = pd.DataFrame()
dolar3 = pd.DataFrame()
```

```python
selic3["SELIC"] = nasdaqdatalink.get("BCB/432", start_date="2015-01-01", collapse='daily')
ipca3["IPCA"] = nasdaqdatalink.get("BCB/13522", start_date='2015-01-01', collapse='daily')
incc3["INCC"] = nasdaqdatalink.get("BCB/192", start_date='2015-01-01', collapse='daily')
pib3["PIB"] = nasdaqdatalink.get("BCB/4380", start_date='2015-01-01', collapse='daily')
confianca_consumidor3["CONFIANCA"] = nasdaqdatalink.get("BCB/4393", start_date='2015-01-01', collapse='daily')
dolar3["DOLAR"] = nasdaqdatalink.get("BCB/10813", start_date='2015-01-01', collapse='daily')
```

```python
macros3 = pd.concat([selic3, confianca_consumidor3, pib3, ipca3, incc3, dolar3], axis=1)
macros3
```

```python
macros3.fillna(method='ffill', inplace=True)
macros3
```

```python
fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])
fig.add_trace(go.Scatter(name="BRCR11.SA", x=comparativo_fiis.index, y=comparativo_fiis["BRCR11.SA"]), row=1, col=1)
fig.add_trace(go.Scatter(name="DOLAR", x=comparativo_fiis.index, y=comparativo_fiis["DOLAR"]), row=1, col=1, secondary_y=True)
fig.update_layout(title_text='BRCR11 VS DOLAR', template='simple_white')
fig.show()
```

```python
fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])
fig.add_trace(go.Scatter(name="HGLG11.SA", x=comparativo_fiis.index, y=comparativo_fiis["HGLG11.SA"]), row=1, col=1)
fig.add_trace(go.Scatter(name="DOLAR", x=comparativo_fiis.index, y=comparativo_fiis["DOLAR"]), row=1, col=1, secondary_y=True)
fig.update_layout(title_text='HGLG11 VS DOLAR', template='simple_white')
fig.show()
```

```python
def heatmap(df):
  comparativo_df = pd.merge(df,macros3, how='inner', left_index=True, right_index=True)
  tabela_df=comparativo_df.corr()
  lista_df = tabela_df.columns

  sns.heatmap(tabela_df, cmap='coolwarm',
              center=0,
              annot=True,
              fmt='.1g',
              xticklabels=lista_df,
              yticklabels=lista_df);
```

```python
#Indice
ibov = pd.DataFrame()
ibov['IBOV'] = yf.download('^BVSP', start='2021-01-01', end='2022-12-31')['Adj Close']

comparativo_ibov = pd.merge(ibov,macros3, how='inner', left_index=True, right_index=True)
tabela_ibov=comparativo_ibov.corr()
lista_ibov = tabela_ibov.columns

sns.heatmap(tabela_ibov, cmap='coolwarm',
            center=0,
            annot=True,
            fmt='.1g',
            xticklabels=lista_ibov,
            yticklabels=lista_ibov);
```

```python
fig = make_subplots(rows=1, cols=1, specs=[[{'secondary_y': True}]])
fig.add_trace(go.Scatter(name='IBOV', x=comparativo_ibov.index, y=comparativo_ibov['IBOV']),row=1, col=1)
fig.add_trace(go.Scatter(name='SELIC', x=comparativo_ibov.index, y=comparativo_ibov['SELIC']),row=1, col=1, secondary_y=True)
fig.update_layout(title_text='IBOV VS SELIC', template='simple_white')
fig.show()
```
